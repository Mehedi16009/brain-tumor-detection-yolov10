```python
import scipy.io
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
```

```python
# Define input and output paths
input_folder = "/content/drive/MyDrive/Brain tumor Dataset/3064 T1-weighted contrast-enhanced MRI (CE-MRI) slices Dataset/data"  # Change to your path
output_folder = "/content/drive/MyDrive/Brain tumor Dataset"

# Create output directory if not exists
os.makedirs(output_folder, exist_ok=True)

# Process each .mat file in the folder
for filename in os.listdir(input_folder):
    if filename.endswith(".mat"):
        mat_path = os.path.join(input_folder, filename)

        # Load the .mat file
        mat_data = scipy.io.loadmat(mat_path)

        # Extract image data (assuming 'img' is the key, change if needed)
        for key in mat_data:
            if isinstance(mat_data[key], np.ndarray):
                img_array = mat_data[key]
                break

        # Normalize and convert image to uint8 format
        img_array = np.array(img_array, dtype=np.float32)
        img_array -= img_array.min()  # Normalize
        img_array /= img_array.max()  # Normalize
        img_array *= 255
        img_array = img_array.astype(np.uint8)

        # Save as JPG
        output_filename = os.path.splitext(filename)[0] + ".jpg"
        output_path = os.path.join(output_folder, output_filename)

        # Convert grayscale to 3-channel (if needed)
        if len(img_array.shape) == 2:
            img_array = cv2.cvtColor(img_array, cv2.COLOR_GRAY2BGR)

        cv2.imwrite(output_path, img_array)
        print(f"Saved: {output_path}")
```

```python
import h5py
import numpy as np
import cv2
import os

# Define input and output paths
input_folder = "/content/drive/MyDrive/Brain tumor Dataset/3064 T1-weighted contrast-enhanced MRI (CE-MRI) slices Dataset/data"  # Change to your path
output_base_folder = "/content/drive/MyDrive/jpg_images"

# Create base output directory if it doesn't exist
os.makedirs(output_base_folder, exist_ok=True)

# Label mapping based on cjdata.label
label_mapping = {
    1: 'meningioma',
    2: 'glioma',
    3: 'pituitary_tumor'
}

# Process each .mat file in the folder
for filename in os.listdir(input_folder):
    if filename.endswith(".mat"):
        mat_path = os.path.join(input_folder, filename)

        # Load the .mat file
        with h5py.File(mat_path, 'r') as mat_data:
            # Access the 'cjdata' group
            if 'cjdata' in mat_data:
                cjdata_group = mat_data['cjdata']

                # Extract the image data
                if 'image' in cjdata_group:
                    img_array = np.array(cjdata_group['image']).T  # Transpose if needed
                else:
                    print(f"'image' dataset not found in 'cjdata' of {filename}.")
                    continue

                # Extract the label
                if 'label' in cjdata_group:
                    label = int(np.array(cjdata_group['label'])[0][0])
                    label_name = label_mapping.get(label, 'unknown')
                else:
                    print(f"'label' dataset not found in 'cjdata' of {filename}.")
                    continue
            else:
                print(f"'cjdata' group not found in {filename}.")
                continue

        # Normalize and convert image to uint8 format
        img_array = img_array.astype(np.float32)
        img_array -= img_array.min()  # Normalize
        img_array /= img_array.max()  # Normalize
        img_array *= 255
        img_array = img_array.astype(np.uint8)

        # Define output directory based on label
        output_folder = os.path.join(output_base_folder, label_name)
        os.makedirs(output_folder, exist_ok=True)

        # Save as JPG
        output_filename = os.path.splitext(filename)[0] + ".jpg"
        output_path = os.path.join(output_folder, output_filename)

        # Convert grayscale to 3-channel (if needed)
        if len(img_array.shape) == 2:
            img_array = cv2.cvtColor(img_array, cv2.COLOR_GRAY2BGR)

        cv2.imwrite(output_path, img_array)
        print(f"Saved: {output_path}")
```
